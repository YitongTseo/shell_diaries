```python
import pandas as pd 
df = pd.read_csv('refmet.csv')
df
```

```python
# Remove duplicates based on the 'formula' column
df_dedup = df.drop_duplicates(subset=['formula'])

df_dedup[['refmet_name','formula']].to_csv("deduped_naive_compound_list.csv",index=False)
```

```python
df[['refmet_name', 'formula']].to_csv('naive_compound_list.csv', index=False)
```

```python
compound_list = list(set([x for x in df['formula'].to_list() if isinstance(x, str) and len(x) > 1]))

import csv

# Save as one row
with open('naive_compound_list.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    for item in compound_list:
        writer.writerow([item ])

```
